```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from sklearn.datasets import make_blobs
```

```python
# Step 1: Generate synthetic data
# Create 4 clusters with 100 points each
data, _ = make_blobs(n_samples=100, centers=4, cluster_std=1.0, random_state=42)

# Plot the raw data
plt.scatter(data[:, 0], data[:, 1], c='blue', s=50)
plt.title("Raw Data")
plt.show()

# Step 2: Compute the distance matrix and perform hierarchical clustering
# Use 'ward' linkage for minimizing variance within clusters
linked = linkage(data, method='ward')

# Step 3: Visualize the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title("Dendrogram")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()

# Step 4: Extract clusters at a specific level
# Decide on a distance threshold to cut the dendrogram
threshold = 7  # This value depends on the dendrogram's structure
clusters = fcluster(linked, threshold, criterion='distance')

# Step 5: Visualize the final clusters
plt.scatter(data[:, 0], data[:, 1], c=clusters, cmap='rainbow', s=50)
plt.title("Hierarchical Clustering Results")
plt.show()

# Print the cluster assignments
print("Cluster Assignments:", clusters)
```
