```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn import set_config

import plotly.express as px
```

```python
df = pd.read_csv('/mnt/Linux/Learning/Python/Datasets/titanic-training-data.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
X = df[['Pclass', 'Age', 'SibSp',
       'Parch', 'Fare']]

X.dropna(inplace=True)

y = df['Survived']
# y = df['Survived']
```

```python
scaler = StandardScaler()

set_config(transform_output='pandas')

scaling_X = scaler.fit_transform(X)
scaling_X
```

```python
pca = PCA()

X_pca = pca.fit_transform(scaling_X)
```

```python
X_pca
```

```python
fig = px.scatter_3d(X_pca, x='pca0', y='pca1', z='pca2', color=y[:X.shape[0]])
fig.show()
```
