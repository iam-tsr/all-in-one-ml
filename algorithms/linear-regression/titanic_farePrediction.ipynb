```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
```

```python
df = pd.read_csv('/mnt/Linux/Learning/Python/Datasets/titanic-training-data.csv')
df.head()
```

```python
def imp_columns(df):
    return (df
            .loc[:, ['Survived', 'Pclass', 'Age', 'SibSp',
       'Parch', 'Fare']]
            .dropna()
    )

df2 = imp_columns(df)
```

```python
df2
```

```python
X = df2.drop(columns=['Fare'])
y = df2['Fare']
```

```python
lr = LinearRegression()
lr.fit(X, y)
```

```python
lr.score(X, y)
```

```python
xgb = XGBRegressor()
xgb.fit(X, y)
xgb.score(X, y)
```
