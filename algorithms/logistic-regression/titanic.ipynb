```python
import pandas as pd

df = pd.read_csv("../dataset/titanic-training-data.csv")

df.head()
```

```python
df.drop(columns=['Pclass', 'SibSp', 'Parch', 'PassengerId', 'Name', 'Ticket', 'Cabin', 'Embarked'], inplace=True)
```

```python
df.tail()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df2 = df.copy()
```

```python
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()

df2[['Sex']] = df[['Sex']].apply(labelencoder.fit_transform)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df2[['Age', 'Fare']] = scaler.fit_transform(df2[['Age', 'Fare']])
df2.head()
```

```python
df2.describe()
```

```python
from sklearn.model_selection import train_test_split

X = df2.drop(columns=['Survived'])
y = df2['Survived']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=128)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```
