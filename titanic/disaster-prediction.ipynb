```python
female_survived = train_df[train_df.Sex == 'female']['Survived']
female_survived_rate = sum(female_survived)/len(female_survived)

print(f"Female who survived: {female_survived_rate:.2f}%")
```

```python
male_survived = train_df[train_df.Sex == 'male']['Survived']
male_survived_rate = sum(male_survived)/len(male_survived)

print(f"Male who survived: {male_survived_rate:.2f}%")
```

```python
from sklearn.ensemble import RandomForestClassifier

features = ["Pclass", "Sex", "SibSp", "Parch"]

X = pd.get_dummies(train_df[features])
y = train_df["Survived"]

model = RandomForestClassifier(n_estimators=100, max_depth=1, random_state=42)
model.fit(X, y)
```

```python
X_test = pd.get_dummies(test_df[features])
predictions = model.predict(X_test)
```

```python
predictions[:5]
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(gender_sub_df['Survived'], predictions)
print(f"Accuracy: {acc*100:.2f}%")
```

```python
from sklearn.tree import plot_tree

# Plot the decision tree
plt.figure(figsize=(10,5))
plot_tree(model.estimators_[0], feature_names=X.columns, class_names=['Not Survived', 'Survived'], filled=True)
plt.show()
```

```python
survived = []
notSurvived = []

for i in range(len(predictions)):
    if predictions[i] == 0:
        survived.append(0)

    elif predictions[i] == 1:
        notSurvived.append(1)
```

```python
# Count the number of survived and not survived
survived_count = len(survived)
not_survived_count = len(notSurvived)

# Plot the data
labels = ['Survived', 'Not Survived']
counts = [survived_count, not_survived_count]

plt.bar(labels, counts, color=['green', 'red'])
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.title('Survived vs Not Survived')
plt.show()
```
